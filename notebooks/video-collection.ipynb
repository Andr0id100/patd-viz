```python
import pandas as pd
import youtube_dl
```

```python
df = pd.read_csv('./video-list.csv')
```

```python
df.head()
```

```python
for index in range(len(df)):
    link = df['link'][index]
    ydl_config = {'outtmpl': f"videos/{index:02d}.mp4"}
    with youtube_dl.YoutubeDL(ydl_config) as ydl:
        ydl.download([link])
```
