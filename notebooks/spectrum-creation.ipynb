```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image, ImageDraw
```

```python
df = pd.read_csv("video-list.csv")
```

```python
df
```

```python
for video_name in range(len(df)):
    print(f"Processing video: {video_name:02d}")
    frame_count = len(os.listdir(f"frames/{video_name:02d}"))
    
    colors = []
    for index in range(1, frame_count+1):
        file_path = f"frames/{video_name:02d}/{index:03d}.png"
        img = plt.imread(file_path)
        color = tuple(np.array(img.mean(axis=0).mean(axis=0) * 255, dtype=int))
        colors.append(color)
```
