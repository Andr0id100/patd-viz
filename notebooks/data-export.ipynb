```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
import json
```

```python
df = pd.read_csv("video-list.csv")
```

```python
data = []
for video_name in range(len(df)):
    print(f"Processing video: {video_name:02d}", df['name'][video_name])
    frame_count = len(os.listdir(f"frames/{video_name:02d}"))
    
    colors = []
    for index in range(1, frame_count+1):
        file_path = f"frames/{video_name:02d}/{index:03d}.png"
        img = plt.imread(file_path)
        color = np.array(img.mean(axis=0).mean(axis=0) * 255, dtype=int).tolist()
        colors.append(color)
        
    video = {}
    video['name'] = df['name'][video_name]
    video['year'] = int(df['year'][video_name])
    video['colors'] = colors
    data.append(video)
```

```python
with open('processed-data.json', 'w') as f:
    json.dump(data, f, indent=4)
```

```python
with open('processed-data.json', 'r') as f:
    data = json.load(f)
```
